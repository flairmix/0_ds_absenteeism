```python
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv('Absenteeism_data.csv')

#drop useless columns
raw_data = raw_data.drop(['ID'], axis=1)

#group "Reason for Absense" for 4 dummies
raw_data['Reason for Absence'] = raw_data['Reason for Absence'].replace([range(0, 15)], 1)
raw_data['Reason for Absence'] = raw_data['Reason for Absence'].replace([15, 16, 17], 2)
raw_data['Reason for Absence'] = raw_data['Reason for Absence'].replace([range(18, 22)], 3)
raw_data['Reason for Absence'] = raw_data['Reason for Absence'].replace([range(22, 29)], 4)

raw_data_dummies = pd.get_dummies(raw_data['Reason for Absence'])

#add dummies in DataFrame
raw_data["Group_1"], raw_data["Group_2"], raw_data["Group_3"], raw_data["Group_4"] = raw_data_dummies[1], raw_data_dummies[2], raw_data_dummies[3], raw_data_dummies[4]

#drop useless columns
raw_data = raw_data.drop(['Reason for Absence'], axis=1)

raw_data
```
